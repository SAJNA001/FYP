```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
import time
```

```python
person_list = [
#     #beauty
    "CTilburyMakeup", "davelackie", "temptalia", "arah_berryman1", "xoShaaan", "Lisa_Eldridge", "Jaclynhill", "GlamorableBlog",
    "labelsforlunch", "HeyAprill", "LauraLouMakeup", "JenLuvsReviews", "_chloehx_", "JeffreeStar", "beautylymin",
    "pintsizedtweets", "Beauty_Balm", "emshelx", "MyBeautyBunny", "BeautifulWhims",
    "hudabeauty", "NikkieTutorials", "jamescharles", "shereeneidriss", "teawithmd", "laradevganmd", "doctorly", "tonyyounmd",  "dr.zionko",  "nopainmoregains",
    
]

num_tweets_per_tag = 10000
```

```python
# Keywords
keywords = [ 
    #beauty keywords
    "#makeuptransformation", "#beauty", "#makeuplessons", "#makeupoftheday", "#1minutemakeup", "#glowingskin", "#clearskin", "#hairgoals", "#hairproducts", "#dailyskincare", 
    "#skincare", "#facecare", "#skincareluxury", "#skinlove", "#skinroutine", "#skincareshop","#skincareroutines","#skintips","#skincareroutines","#makeup",
    "#skincareproducts","#newskincare","#dermalogica", "#cosmetics", "#beautybasics", "#skincaretips","#beautytrends","#Makeuptips",
]
```

```python
def scrape_data_user(twitter_user, start_date="2019-01-01", end_date="2022-07-31", num_tweets_per_tag=100):
    df = pd.DataFrame()
    start = time.time()  
    for word in person_list:
        try:
            df = df.append(pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
                f'{word} from:{twitter_user} lang:en since:{start_date} until:{end_date}').get_items(), num_tweets_per_tag)))
        except Exception as e:
            print("An error occured.\n")
            continue
    
    print(f'Number of tweets from {twitter_user} : {df.shape}')
    print(df.columns)
    
    
    df_ = None
    if df.shape[0] > 0:
        df["username"] = df["user"].apply(lambda x: x["username"])
        df_ = df[["id", "url", "date", "content", "username", "replyCount", "retweetCount", "likeCount", "quoteCount", "hashtags"]]
        
    
    print(f'Time taken : {((time.time() - start)/60):.1f} mins')
    return df_
```

```python
# Initializing Dictionary of DataFrames for Each of the 23 Countries
personsDf = {}
```

```python
final_df = pd.DataFrame()
for person in person_list:
    personsDf[person] = scrape_data_user(person)
    final_df = final_df.append(personsDf[person])
    
final_df = pd.DataFrame(final_df)
```

```python
final_df = final_df.drop_duplicates(subset=["id"])
print(final_df.groupby("username").size())
```

```python
# Cleaning Data
df_indexes_v2 = []
user_dict = {}
for i in range(len(final_df)):
    tweet = final_df["content"].iloc[i]
    
    # To remove tweets that have more hashtags than normal text
    word_list = tweet.lower().split()
    num_normal = 0
    num_tags = 0
    for j in range(len(word_list)):
        temp = word_list[j]
        if temp[0] == '#':
            num_tags += 1
        else:
            num_normal += 1
    if num_tags < num_normal:
        df_indexes_v2.append(i)
    

df_v2 = final_df.iloc[df_indexes_v2]

#Remove tweets that have 0 likes, retweets, or comments
# df_v2 = df_v2.drop(df_v2[(df_v2["likeCount"] == 0) | (df_v2["replyCount"] == 0) | (df_v2["retweetCount"] == 0)].index)
df_v2 = df_v2[(df_v2["retweetCount"] > 0) & (df_v2["replyCount"] > 0) & (df_v2["likeCount"] > 0)]
# df_v2 = df_v2.iloc[indices]
print(f'Shape of df after cleaning: {df_v2.shape}')
# print(len(indices))
```
